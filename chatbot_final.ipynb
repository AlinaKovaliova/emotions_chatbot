```python
# !pip install transformers torch deep-translator

from transformers import pipeline
from deep_translator import GoogleTranslator
import pandas as pd

df = pd.read_csv(r"C:\Users\koval\combined_emotion.csv")

df = df[['sentence', 'emotion']]

emotion_classifier = pipeline("text-classification", model="bhadresh-savani/distilbert-base-uncased-emotion", top_k=None)

sample_texts = df.sample(5)['sentence'].tolist()
predictions = emotion_classifier(sample_texts)

def translate_to_english(text):
    return GoogleTranslator(source='auto', target='en').translate(text)

def chat_bot_response(user_input):
    translated_text = translate_to_english(user_input)
    emotion = emotion_classifier(translated_text)[0][0]['label']
    
    responses = {
        "sadness": "Я чувствую, что тебе грустно. Давай послушаем что-то весёлое! 🎵 Вот плейлист для поднятия настроения: https://open.spotify.com/playlist/37i9dQZF1DX3rxVfibe1L0",
        "joy": "Отличное настроение! Как насчёт активного отдыха? 🚴‍♂️🏕️ Попробуй покататься на велосипеде, отправиться на прогулку или заняться спортом!",
        "anger": f"Я понимаю, что тебя раздражает {user_input}. Давай попробуем разобраться в ситуации и найти решение. Попробуй сделать несколько глубоких вдохов! 🌿",
        "fear": "Ты испытываешь страх. Главное – сохранять спокойствие. Всё будет хорошо! 😊 Не забывай, что страхи часто исчезают, когда мы о них говорим!",
        "love": "Ты чувствуешь любовь! Это замечательно! ❤️ Любовь – это одно из самых прекрасных чувств!",
        "surprise": "Ты удивлён! Пусть этот день принесёт тебе только приятные сюрпризы! 🎉 Наслаждайся моментом!"
    }

    return f"{responses.get(emotion, 'Интересная эмоция! Давай обсудим её.')}"

if __name__ == "__main__":
    while True:
        user_input = input("Введите ваше сообщение: ")
        if user_input.lower() in ["выход", "exit"]:
            break
        print(chat_bot_response(user_input))
```
